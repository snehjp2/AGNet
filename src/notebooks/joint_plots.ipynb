```python
import pandas as pd
from scipy import stats
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
from math import sqrt
from sklearn.metrics import mean_squared_error, r2_score
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
df = pd.read_csv('/Users/SnehPandya/Desktop/z_results_april.csv')
df_X = df['ground_truth']
df_Y = df['network_predictions']
```

```python
sns.set(style='white')
plot = sns.jointplot(df_X, df_Y, color='darkred', s=10, alpha=.5)
plot.ax_joint.plot(df_X,df_X, color='black', linewidth = 1, label = 'Mass Ground Truth')
plot.ax_joint.set_xlabel('Redshift Ground Truth', fontsize=16)
plot.ax_joint.set_ylabel('Redshift Prediction',fontsize=16)
# plot.ax_marg_x.set_xlim(7,10.5)
# plot.ax_marg_y.set_ylim(7,10.5)
plot.ax_joint.annotate('Pearson = 0.875' +'\n$R^2$ score = 0.759' + '\nRMSE = 0.365', xy=(.1, .85),
                 xycoords='axes fraction', size=13)
plt.show()
# plot.savefig('/Users/SnehPandya/Desktop/z_results_april.pdf', format='pdf')
```
